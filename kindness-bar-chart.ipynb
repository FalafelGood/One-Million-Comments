```python
# AMDG
import os
import json

channel_ids = []
kindness_scores = []
volatility_scores = []
total_comments = 0

with os.scandir('channel-ratings') as files:
    for idx, file in enumerate(files):
        with open(file, 'r') as open_file:
            data = json.load(open_file)
            channel_ids.append(data['channel-id'])
            kindness_scores.append(data['kindness'])
            volatility_scores.append(data['volatility'])
            total_comments += data['num-comments-analyzed']
```

```python
sorted_data = sorted(zip(kindness_scores, channel_ids), key=lambda x: x[0])
sorted_kindness = [score for score, _ in sorted_data]
channels_by_kindness = [channel_id for _, channel_id in sorted_data]
```

```python
channels_by_kindness
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure with good size
plt.figure(figsize=(12, max(8, len(channels_by_kindness) * 0.3)))

# Create horizontal bar chart (better for many labels)
bars = plt.barh(range(len(channels_by_kindness)), sorted_kindness, 
                color=plt.cm.viridis(np.linspace(0.2, 0.8, len(sorted_kindness))))

# Set y-axis labels to channel names
plt.yticks(range(len(channels_by_kindness)), channels_by_kindness, fontsize=9)

# Labels and title
plt.xlabel('Kindness Score', fontsize=12, fontweight='bold')
plt.ylabel('Channel', fontsize=12, fontweight='bold')
plt.title('Kindness Scores by Channel', fontsize=14, fontweight='bold', pad=20)

# Add grid for better readability
plt.grid(axis='x', alpha=0.3, linestyle='--')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
